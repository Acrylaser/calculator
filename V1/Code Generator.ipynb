```python
import pandas as pd

# Read the CSV file
data = pd.read_csv("output_table.csv", encoding="utf-8")

# Initialize dictionaries to store the data
categories = {}
materials = {}
prices = {}

# Process the data and populate the dictionaries
for _, row in data.iterrows():
    category = row["Category"]
    material = row["Material"]
    calliper = row["Calliper"]
    price = row["Price"]
    
    # Populate the categories dictionary
    if category not in categories:
        categories[category] = []
    if material not in categories[category]:
        categories[category].append(material)
    
    # Populate the materials dictionary
    if material not in materials:
        materials[material] = []
    if calliper not in materials[material]:
        materials[material].append(calliper)
    
    # Populate the prices dictionary
    if category not in prices:
        prices[category] = {}
    if material not in prices[category]:
        prices[category][material] = {}
    prices[category][material][calliper] = price

# Generate the JavaScript code
js_code = "const categories = {\n"
for category, material_list in categories.items():
    js_code += f'  "{category}": {str(material_list)},\n'
js_code += "};\n\n"

js_code += "const materials = {\n"
for material, calliper_list in materials.items():
    js_code += f'  "{material}": {str(calliper_list)},\n'
js_code += "};\n\n"

js_code += "const prices = {\n"
for category, material_dict in prices.items():
    js_code += f'  "{category}": {{\n'
    for material, calliper_dict in material_dict.items():
        js_code += f'    "{material}": {{\n'
        for calliper, price in calliper_dict.items():
            js_code += f'      "{calliper}": {price},\n'
        js_code += "    },\n"
    js_code += "  },\n"
js_code += "};"

# Print or save the JavaScript code
print(js_code)
# If you want to save the code to a file, you can use the following line:
# with open("output.js", "w", encoding="utf-8") as file:
#     file.write(js_code)
```
